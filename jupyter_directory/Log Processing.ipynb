```python
import re
import matplotlib.pyplot as plt
import numpy as np
text = ""
scores = ""

def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n

logs = [
    "../output/nell/afe4_top25%_NELL995_new_SL/log.txt",
    "../output/nell/0b41_hilrtop25%_NELL995_new_SL/log.txt",
    "../output/nell/2d79_ignore75%_NELL995_new_SL/log.txt",
    "../output/nell/9c05_ignore50%_NELL995_new_SL/log.txt",
    "../output/nell/ec8e_ignore25%_NELL995_new_SL/log.txt",
]

scores = [
    "../output/nell/afe4_top25%_NELL995_new_SL/scores.txt",
    "../output/nell/0b41_hilrtop25%_NELL995_new_SL/scores.txt",
    "../output/nell/2d79_ignore75%_NELL995_new_SL/scores.txt",
    "../output/nell/9c05_ignore50%_NELL995_new_SL/scores.txt",
    "../output/nell/ec8e_ignore25%_NELL995_new_SL/scores.txt",
]

# set this to switch logs
trialnum = 0

with open(logs[trialnum], "r") as logfile:
    text = logfile.read()
    
with open(scores[trialnum], "r") as scorefile:
    scores = scorefile.read()
    scores = scores.split("\n\n")
    
for index in range(len(scores)):
    scores[index] = np.array([float(x) for x in re.findall("[0-9]*\.[0-9]*", scores[index])])
    
text = text[255:]
text = re.sub('.*\/.*\/.*\[ (?!RL).*\n', '', text)
text = text.split("\n")
datablocks = []
print(len(text))
if trialnum == 0:
    for i in range(1, 41):
        datablocks.append('\n'.join(text[(i - 1)*2000 : i*2000]))
else:
    for i in range(1, 21):
        if i%2 != 0:
            even = ((i - 1)/2)*1500 # even corresponds to SL
            odd = (((i - 1)/2) + 1)*1000 # odd corresponds to RL
            base = int(even + odd - 1000)
            end = int(even + odd)
#             print("[Odd] | base: {}\tend: {}".format(base, end))
            datablocks.append('\n'.join(text[base : end]))
        else:
            even = (i/2)*1500
            odd = (i/2)*1000
            base = int(even + odd - 1500)
            end = int(even + odd)
#             print("[Even] | base: {}\tend: {}".format(base, end))
            datablocks.append('\n'.join(text[base : end]))

for count, value in enumerate(datablocks):
    datablocks[count] = re.sub(".*\/.*avg_ep_correct  ", "", value)
    datablocks[count] = re.sub(", .*\]", ",", datablocks[count])
    datablocks[count] = re.sub("\n", "", datablocks[count])
    datablocks[count] = datablocks[count][:-1].split(',')
    datablocks[count] = [float(x) for x in datablocks[count]]

groups = []
for i in range(1, len(datablocks), 2):
    groups.append(datablocks[i])
    
for count, group in enumerate(groups):
    group = moving_average(group, n=50)
    plt.plot(list(range(len(group))), group)
    plt.show()
    if count == 0:
        print("Hits@1: {}\nHits@3: {}\nHits@5: {}\nHits@10: {}\nHits@20: {}\nauc: {}".format(scores[count][0], scores[count][1], scores[count][2], scores[count][3], scores[count][4], scores[count][5]))
    elif trialnum == 0 and (count > 0 and count < 19):
        diff = np.divide(np.subtract(scores[count], scores[count - 1]), scores[count - 1])*100
        print("Hits@1: {}\nHits@3: {}\nHits@5: {}\nHits@10: {}\nHits@20: {}\nauc: {}".format(diff[0], diff[1], diff[2], diff[3], diff[4], diff[5]))
    elif trialnum != 0 and (count > 0 and count < 9):
        diff = np.divide(np.subtract(scores[count], scores[count - 1]), scores[count - 1])*100
        print("Hits@1: {}\nHits@3: {}\nHits@5: {}\nHits@10: {}\nHits@20: {}\nauc: {}".format(diff[0], diff[1], diff[2], diff[3], diff[4], diff[5]))
    plt.clf()
```
