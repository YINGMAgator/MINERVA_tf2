```python
# test the problems with exploration
class Node():
    def __init__(self, depth, value):
        self.count = 0
        self.depth = depth
        self.value = value
        self.children = {}
        
    def add_child(self, depth, value):
        self.children[value] = Node(depth, value)
        return self.children[value]
        
    def remove_child(self, node):
        try:
            self.children.remove(node)
        except:
            print("child not found")
    
class Trie():
    def __init__(self):
        self.root = Node(0, "root")
        self.depth = 0
        self.nodes = 1
        
    def add_path(self, path):
        if len(path) != 4:
            return
        traverse = self.root
        depth = 1
        for node in path:
            try:
                traverse = traverse.children[node]
            except:
                self.nodes += 1
                if (depth > self.depth):
                    self.depth += 1
                traverse = traverse.add_child(depth, node)
            depth += 1
        traverse.count += 1
        
    def visualize(self):
        self.print_arr = [0,0,0,0,0]
        self.update_lists(self.root)
        return self.print_arr
                
    def update_lists(self, node):
        for x in list(node.children.keys()):
            self.update_lists(node.children[x])
            try:
                self.print_arr[node.depth + 1] += 1
            except:
                self.print_arr[node.depth + 1] = 1

import random
from operator import add
averages_origin = [0,0,0,0,0]
averages_nosl = [0,0,0,0,0]
averages_slrl = [0,0,0,0,0]
for query in range(len(queries)):
    orig_reward_nodes = Trie()
    for path in orig_reward_queries[list(orig_reward_queries.keys())[query]]:
        orig_reward_nodes.add_path(path.split('\n')[0].split('	'))
    averages_origin = list(map(add, averages_origin, orig_reward_nodes.visualize()))
    
    no_sl_nodes = Trie()
    for path in no_sl_queries[list(no_sl_queries.keys())[query]]:
        no_sl_nodes.add_path(path.split('\n')[0].split('	'))
    averages_nosl = list(map(add, averages_nosl, no_sl_nodes.visualize()))
    
    sl_rl_nodes = Trie()
    for path in sl_rl_queries[list(sl_rl_queries.keys())[query]]:
        sl_rl_nodes.add_path(path.split('\n')[0].split('	'))
    averages_slrl = list(map(add, averages_slrl, sl_rl_nodes.visualize()))
    
averages_origin = [x/len(queries) for x in averages_origin]
averages_nosl = [x/len(queries) for x in averages_nosl]
averages_slrl = [x/len(queries) for x in averages_slrl]
print("AVERAGES")
print(averages_origin)
print(averages_nosl)
print(averages_slrl)
```
